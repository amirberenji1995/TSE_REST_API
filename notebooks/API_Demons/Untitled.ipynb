```python
import requests
import pandas as pd
import datetime
```

## Getting the List of Available Symbols

```python
symbols_url = 'http://127.0.0.1:8000/symbols'
list_of_symbols = requests.get(symbols_url).json()
```

```python
list_of_symbols
```

## Recieving Detailed Information (With Declaration of the Duration)

```python
url = symbols_url + '/' + list_of_symbols['1']
headers = {"starting": "2019-06-26", 'ending': '2020-06-10'}
r = requests.get(url, headers=headers)
```

```python
detailed_symbol = r.json()
```

### Creating a DataFrame from the JSON Response

```python
df = pd.read_json(detailed_symbol)
```

```python
df
```

## Recieving Detailed Information (No Duration Declared --> Receving the Whole)

```python
url = symbols_url + '/' + list_of_symbols['1']
rr = requests.get(url)
```

```python
detailed_symbol = rr.json()
```

```python
df = pd.read_json(detailed_symbol)
```
